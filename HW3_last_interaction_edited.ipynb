```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('attribution_allocation_student_data.csv')
df
```

```python
input_file = pd.read_csv('channel_spend_student_data.csv', skiprows=1, names = ['tier', 'spend'])
input_file
```

```python
input_file['spend']=input_file['spend'].apply(ast.literal_eval)
```

# calculate cost for each channel

```python
tier1_spend_dict = input_file.iloc[0,1]
tier2_spend_dict = input_file.iloc[1,1]
tier3_spend_dict = input_file.iloc[2,1]
```

```python
tier1_spend = pd.DataFrame(list(tier1_spend_dict.items()),columns = ['channel','spend'])
tier2_spend = pd.DataFrame(list(tier2_spend_dict.items()),columns = ['channel','spend'])
tier3_spend = pd.DataFrame(list(tier3_spend_dict.items()),columns = ['channel','spend'])
```

```python
tier1 = df[df['tier'] == 1]
tier2 = df[df['tier'] == 2]
tier3 = df[df['tier'] == 3]
```

# Last Interaction

```python
convert1 = tier1[tier1['convert_TF'] == True]
convert2 = tier2[tier2['convert_TF'] == True]
convert3 = tier3[tier3['convert_TF'] == True]
print(convert1)
print(convert2)
print(convert3)
```

```python
convert1_combined = convert1.iloc[:,1:6]
convert1_combined['combined']= convert1_combined.values.tolist()
convert1_combined.drop(convert1_combined.iloc[:, 0:5], inplace=True, axis=1)
convert1_combined
```

```python
convert2_combined = convert2.iloc[:,1:6]
convert2_combined['combined']= convert2_combined.values.tolist()
convert2_combined.drop(convert2_combined.iloc[:, 0:5], inplace=True, axis=1)
convert2_combined
```

```python
convert3_combined = convert3.iloc[:,1:6]
convert3_combined['combined']= convert3_combined.values.tolist()
convert3_combined.drop(convert3_combined.iloc[:, 0:5], inplace=True, axis=1)
convert3_combined
```

```python
touch_list1 = convert1_combined['combined'].dropna().tolist()
for i in range(len(touch_list1)):
    touch_list1[i] = [x for x in touch_list1[i] if str(x) != 'nan']
touch_list1
```

```python
touch_list2 = convert2_combined['combined'].dropna().tolist()
for i in range(len(touch_list2)):
    touch_list2[i] = [x for x in touch_list2[i] if str(x) != 'nan']
touch_list2
```

```python
touch_list3 = convert3_combined['combined'].dropna().tolist()
for i in range(len(touch_list3)):
    touch_list3[i] = [x for x in touch_list3[i] if str(x) != 'nan']
touch_list3
```

```python
counts_1 = dict()
counts_2 = dict()
counts_3 = dict()

for i in range(len(touch_list1)):
    counts_1[touch_list1[i][-1]] = counts_1.get(touch_list1[i][-1],0)+ 1
    
for i in range(len(touch_list2)):
    counts_2[touch_list2[i][-1]] = counts_2.get(touch_list2[i][-1],0)+ 1

for i in range(len(touch_list3)):
    counts_3[touch_list3[i][-1]] = counts_3.get(touch_list3[i][-1],0)+ 1

print(counts_1)
print(counts_2)
print(counts_3)
```

```python
channel_df1 = pd.DataFrame(list(counts_1.items()),columns = ['channel','converts'])
channel_df1
```

```python
channel_df2 = pd.DataFrame(list(counts_2.items()),columns = ['channel','converts'])
channel_df2
```

```python
channel_df3 = pd.DataFrame(list(counts_3.items()),columns = ['channel','converts'])
channel_df3
```

```python
last_interaction_cac1 = pd.merge(channel_df1, tier1_spend, how='inner', right_on=['channel'],left_on=['channel'])
last_interaction_cac1['cac'] = last_interaction_cac1['spend']/last_interaction_cac1['converts']
last_interaction_cac1
```

```python
last_interaction_cac2 = pd.merge(channel_df2, tier2_spend, how='inner', right_on=['channel'],left_on=['channel'])
last_interaction_cac2['cac'] = last_interaction_cac2['spend']/last_interaction_cac2['converts']
last_interaction_cac2
```

```python
last_interaction_cac3 = pd.merge(channel_df3, tier3_spend, how='inner', right_on=['channel'],left_on=['channel'])
last_interaction_cac3['cac'] = last_interaction_cac3['spend']/last_interaction_cac3['converts']
last_interaction_cac3
```

```python
last_interaction_cac = pd.DataFrame()
last_interaction_cac['channel'] = last_interaction_cac1['channel']
last_interaction_cac['converts'] = last_interaction_cac1['converts'] + last_interaction_cac2['converts'] + last_interaction_cac3['converts']
last_interaction_cac['spend'] = last_interaction_cac1['spend'] + last_interaction_cac2['spend'] + last_interaction_cac3['spend']
last_interaction_cac['cac'] = last_interaction_cac['spend']/last_interaction_cac['converts']
last_interaction_cac
```
